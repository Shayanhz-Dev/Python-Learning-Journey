# Data Types for ndarrays

```python
import numpy as np
```

### Casting Data Types
Use the `astype` method to explicitly convert or cast an array from one `dtype` to another:


```python
arr1 = np.array([1, 2, 3], dtype=np.float64)
```

```python
arr2 = np.array([1, 2, 3], dtype=np.int32)
```

```python
arr1.dtype
```

```python
arr2.dtype
```

```python
arr = np.array([1, 2, 3, 4, 5])
```

```python
arr.dtype
```

```python
float_arr = arr.astype(np.float64)
```

```python
float_arr
```

```python
float_arr.dtype
```

```python
arr = np.array([3.7, -1.2, -2.6, 0.5, 12.9, 10.1])
```

```python
arr
```

#### Floating-point to Integer
When casting floats to integers, the decimal part is **truncated** (not rounded):


```python
arr.astype(np.int32)
```

#### String to Numeric
Arrays containing strings that represent numbers can be easily converted to numeric formats using `astype`:


```python
numeric_strings = np.array(["1.25", "-9.6", "42"], dtype=str)
```

```python
numeric_strings.astype(float)
```

```python
int_array = np.arange(10)
```

```python
calibers = np.array([.22, .270, .357, .380, .44, .50], dtype=np.float64)
```

#### Casting via Another Array's `dtype`
You can dynamically cast an array by passing the `dtype` attribute of another existing array:


```python
int_array.astype(calibers.dtype)
```

#### Shorthand Type Codes
NumPy supports concise string codes for specifying data types (e.g., `"u4"` stands for unsigned 32-bit integer):


```python
zeros_uint32 = np.zeros(8, dtype="u4")
```

```python
zeros_uint32
```
