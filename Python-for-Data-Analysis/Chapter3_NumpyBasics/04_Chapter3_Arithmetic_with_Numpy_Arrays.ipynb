# Arithmetic with Numpy Arrays

```python
import numpy as np
```

```python
arr = np.array([[1., 2., 3.], [4., 5., 6.]])
```

```python
arr
```

## Element-wise Operations
Arrays are important because they enable you to express batch operations on data without writing any `for` loops. This is usually called **vectorization**. Any arithmetic operations between equal-size arrays apply the operation element-wise.


```python
arr * arr
```

```python
arr - arr
```

## Operations with Scalars
Arithmetic operations with scalars propagate the scalar argument to each element in the array.


```python
 1 / arr
```

```python
arr ** 2
```

```python
arr2 = np.array([[0., 4., 1.], [7., 2., 12.]])
```

```python
arr2
```

## Comparisons Between Arrays
Evaluating comparisons between arrays of the same size yields boolean arrays.


```python
arr2 > arr
```
