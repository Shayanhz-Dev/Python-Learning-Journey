```python
some_dict = {"a": 1, "b": 2, "c": 3}
```

```python
for key in some_dict:
    print(key)
```

```python
dict_iterator = iter(some_dict)
```

```python
list(dict_iterator)
```

```python
def squares(n=10):
    print(f"Generating squares from 1 to {n ** 2}")
    for i in range(1, n + 1):
        yield i ** 2
```

```python
gen = squares()
```

```python
for x in gen:
    print(x, end=" ")
```

```python
def _make_gen():
    for x in range(100):
        yield x ** 2
gen = _make_gen
```

```python
sum(x ** 2 for x in range(100))
```

```python
dict((i, i ** 2) for i in range(5))
```

## itertools module

```python
import itertools
```

```python
def first_letter(x):
    return x[0]
```

```python
names = ["Alan", "Wes", "Will", "Albert", "Steven"]
```

```python
for letter, names in itertools.groupby(names, first_letter):
    print(letter, list(names)) # names is a generator
```
