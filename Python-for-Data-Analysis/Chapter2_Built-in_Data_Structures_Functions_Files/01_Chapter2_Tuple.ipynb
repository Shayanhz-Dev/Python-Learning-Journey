## Unpacking tuples

```python
tup = (4, 5, 6)
```

```python
a, b, c = tup
```

```python
b
```

```python
tup = 4, 5, (6, 7)
```

```python
a, b, (c, d) = tup
```

```python
d
```

```python
a, b = 1, 2
```

```python
a
```

```python
b
```

```python
b, a = a, b
```

```python
a
```

```python
b
```

```python
seq = [(1, 2, 3), (4, 5, 6), (7, 8, 9)]
```

```python
for a, b, c in seq:
    print(f"a={a}, b={b}, c={c}")
```

```python
values = 1, 2, 3, 4, 5
```

```python
a, b, *_ = values
```

```python
a
```

```python
b
```

```python
_
```
