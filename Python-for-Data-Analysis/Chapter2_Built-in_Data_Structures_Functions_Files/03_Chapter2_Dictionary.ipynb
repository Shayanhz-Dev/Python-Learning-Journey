## Default values

```python
words = ['apple', 'bat', 'bar', 'atom', 'book']
```

```python
by_letter = {}
```

```python
for word in words:
    letter = word[0]
    if letter not in by_letter:
        by_letter[letter] = [word]
    else:
        by_letter[letter].append(word)
```

```python
by_letter
```

```python
from collections import defaultdict
```

```python
by_letter = defaultdict(list)
```

```python
for word in words:
    by_letter[word[0]].append(word)
```

```python
by_letter
```
