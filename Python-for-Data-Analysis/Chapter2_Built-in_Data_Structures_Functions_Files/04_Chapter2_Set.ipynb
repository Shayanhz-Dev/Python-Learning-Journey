```python
sorted("horse race")
```

## Zip

```python
seq1 = ["foo", "bar", "baz"]
```

```python
seq2 = ["one", "two", "three"]
```

```python
zipped = zip(seq1, seq2)
```

```python
list(zipped)
```

```python
seq3 = [False, True]
```

```python
list(zip(seq1, seq2, seq3))
```

```python
for index, (a, b) in enumerate(zip(seq1, seq2)):
    print(F"{index}: {a}, {b}")
```
