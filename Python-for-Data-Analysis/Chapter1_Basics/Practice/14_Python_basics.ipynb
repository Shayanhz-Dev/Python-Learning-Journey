## For-loops

```python
sequence = [1, 2, None, 4, None, 5]
total = 0
for value in sequence:
    if value is None:
        continue
    total += value
print(total)
```

```python
sequence = [1, 2, 0, 4, 6, 5, 2, 1]
total_until_5 = 0
for value in sequence:
    if value == 5:
        break
    total_until_5 += value
print(total_until_5)
```

```python
for i in range(4):
    for j in range(4):
        if j > i:
            break
        print((i,j))
```

```python
iterator = [
    (2, 4, 6),
    (7, 8, 9),
    (1, 3, 5)
]
for a, b, c in iterator:
    print(a, b, c)
```
