## Control Flow

```python
x = -5
if x < 0:
    print("it's negative")
```

```python
if x < 0:
    print("Its's negative")
elif x == 0:
    print("Equal to zero")
elif 0 < x < 5:
    print("Positive but smaller than 5")
else:
    print("Positive and larger than or equal to 5")
```
